```python
import pint
import scipy

si = pint.UnitRegistry()
R_str = (scipy.constants.R * si.J / si.mol / si.K) / (28.966 * si.g/si.mol)
```

```python
def rho(p, T):
    return p/R_str/T

print(x:=rho(1000*si.hPa, 300*si.K).to_base_units())
x.to(si.g/si.cm**3)
```

```python
f"{x.to(si.lbs / si.pint)=:.3g}"
```

```python
from matplotlib import pyplot as plt
import numpy as np

si.setup_matplotlib()
plt.plot(T:=np.linspace(200,300) * si.K, rho(1000 * si.hPa, T), label = r'$\rho$')
plt.legend()
plt.gca().xaxis.set_units(si.celsius)
plt.gca().yaxis.set_units(si.kg / si.m**3)
```
